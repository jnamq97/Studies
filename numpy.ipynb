Numpy는 "다차원 배열"을 효과적으로 처리할 수 있음

```python
import numpy as np
array = np.array([1,2,3,4])
print(array)
print(array.size)
print(array.dtype)
```

Broadcasting

```python
import numpy as np
#broadcasting
array1 = np.arange(4).reshape(2,2)
array2 = np.arange(2)
array3 = array1 + array2
print(array3)
```

```python
import numpy as np
#broadcasting
array1 = np.arange(0,8).reshape(2,4)
array2 = np.arange(0,8).reshape(2,4)
array3 = np.concatenate([array1, array2], axis=0)
array4 = np.arange(0,4).reshape(4,1)
print(array3)
print(array4)
print(array3+array4)
```

마스킹

```python
import numpy as np
#마스킹
array1 = np.arange(16).reshape(4,4)
print(array1)
array2 = array1 < 10
print(array2)
array1[array2] = 100
print(array1)
```

Numpy 집계함수

```python
import numpy as np
#numpy 집계함수
array = np.arange(16).reshape(4,4)

print("최대값:", np.max(array))
print("최소값:", np.min(array))
print("합계:", np.sum(array))
print("평균값:", np.mean(array))
print("합계:", np.sum(array, axis=0))
```

단일 객체 저장 및 불러오기

```python
import numpy as np
array = np.arange(0,10)
np.save('saved.npy', array)
result = np.load('saved.npy')
print(result)
```

복수 객체 저장 및 불러오기

```python
import numpy as np
array1 = np.arange(0,10)
array2 = np.arange(11,20)
np.savez('saved.npz', array1=array1, array2=array2)

data = np.load('saved.npz')
print(data['array1'])
print(data['array2'])
```

인덱싱, 정렬

```python
import numpy as np

array = np.array([5,9,10,3,1])
array.sort()
print(array)
print(array[::-1])
```

```python
import numpy as np
#각 열을 기준으로 정렬
array = np.array([[5, 9, 10, 3, 1], [8, 3, 4, 2, 5]])
print(array)
array.sort(axis=0)
print(array)
```

균일한 간격으로 데이터 생성

```python
import numpy as np
#균일한 간격으로 데이터 생성
array = np.linspace(0, 10, 5)
print(array)
```

random.seed

```python
import numpy as np
#난수의 재연 (실행마다 결과 동일)
np.random.seed(7)
print(np.random.randint(0,10, (2,3)))
```

Numpy 배열 객체 복사

```python
#Numpy 배열 객체 복사
array1 = np.arange(0, 10)
array2 = array1.copy()
array2[0] = 99
print(array1)
array3 = array1
array3[0] = 99
print(array1)
```
